```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R_sum_nominal = 20 # мат. ожидание общего сопротивления
n = 100 # количество экспериментов
```

### Последовательное соединение:

```python
R_nominal = R_sum_nominal * 2           # мат. ожидание
sigma = R_nominal * 0.1                 # отклонение
```

```python
R_list = np.round(np.array([np.random.normal(R_nominal, sigma) for _ in range(n)]), decimals=2)
plt.hist(R_list, bins=50, density=True)
plt.title("Распределение номиналов сопротивлений последовательной схемы")
plt.xlabel("Сопротивление (Ом)")
plt.ylabel("Вероятность")
plt.show()
```

```python
R_sum_list = R_list / 2
plt.hist(R_sum_list, bins=50, density=True)
plt.title("Распределение номиналов общих сопротивлений последовательной схемы")
plt.xlabel("Сопротивление (Ом)")
plt.ylabel("Вероятность")
plt.show()
```

```python
mean = np.round(np.mean(R_sum_list), decimals=2)
disp = np.round(np.var(R_sum_list, ddof=0) ** 0.5, decimals=2)
F = sum((R_nominal - R_list) ** 2)
sigma = np.round(((F / (n - 2 + 1)) ** 0.5), decimals=3)
print(f"Мат. ожидание: {mean} Ом")
print(f"Дисперсия: {disp} Ом")
print(f"Сигма: {sigma} Ом")
```

### Параллельное соединение:

```python
R_nominal = R_sum_nominal / 2   # мат. ожидание
sigma = R_nominal * 0.1         # отклонение
```

```python
R_list = np.round(np.array([np.random.normal(R_nominal, sigma) for _ in range(n)]), decimals=2)
plt.hist(R_list, bins=30, density=True)
plt.title("Распределение номиналов сопротивлений параллельной схемы")
plt.xlabel("Сопротивление (Ом)")
plt.ylabel("Вероятность")
plt.show()
```

```python
R_sum_list = R_list * 2
plt.hist(R_sum_list, bins=50, density=True)
plt.title("Распределение номиналов общих сопротивлений параллельной схемы")
plt.xlabel("Сопротивление (Ом)")
plt.ylabel("Вероятность")
plt.show()
```

```python
mean = np.round(np.mean(R_sum_list), decimals=2)
disp = np.round(np.var(R_sum_list, ddof=0) ** 0.5, decimals=2)
F = sum((R_nominal - R_list) ** 2)
sigma = np.round(((F / (n - 2 + 1)) ** 0.5), decimals=3)
print(f"Мат. ожидание: {mean} Ом")
print(f"Дисперсия: {disp} Ом")
print(f"Сигма: {sigma} Ом")
```
