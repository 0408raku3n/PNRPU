```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = []
k_all = []
n_prev = 0
for _ in range(100):
    state = -1
    k = 0
    while True:
        match state:
            case -1:
                r = rd.random()
                if r < 0.8:
                    state = 0
                else:
                    state = -1
            case 0:
                r = rd.random()
                k += 1
                if r < 0.45:
                    state = 0
                elif r < 0.85:
                    state = 1
                else:
                    state = 2
            case 1:
                r = rd.random()
                k += 1
                if r < 0.45:
                    state = 1
                else:
                    state = 2
            case 2:
                k_all.append(k)
                n.append((k + n_prev) / (len(n) + 1))
                n_prev = k + n_prev
                break
print(k_all)
print(n)
```

```python
plt.figure(figsize=(12, 8))

plt.plot(np.arange(1, 101), n, color='skyblue')
# plt.plot(np.arange(1, 101), [n[22] * 0.9] * 100, color='red')
# plt.plot(np.arange(1, 101), [n[22] * 1.1] * 100, color='red')
# plt.plot(np.arange(1, 101), [n[22]] * 100, color='green')

plt.show()
# print(n[22])
# print(n[47])
```
