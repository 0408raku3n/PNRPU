```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numbers = np.array([9083, 4260, 5277, 4998, 4298, 5204, 3965, 4028, 8936, 5148, 
                    1762, 8713, 1189, 1090, 8989, 7273, 3213, 1935, 9321, 4820,
                    2023, 2589, 1740, 424, 8924, 5, 1969, 1636, 7237, 1227,
                    7965, 3855, 4765, 703, 1678, 841, 7543, 308, 9732, 1289,
                    7690, 480, 8098, 9629, 4819, 7219, 7241, 5128, 3853, 1921,
                    9292, 426, 9573, 4903, 5916, 6576, 8368, 3270, 6641, 33,
                    6295, 9765, 1112, 5761, 2575, 6837, 3336, 9322, 7403, 8345])
boy = np.array(numbers // 100 / 100)
girl = np.array(numbers % 100 / 100)
```

```python
N = np.arange(1, len(numbers) + 1)
is_date = [0] * len(numbers)
for i in range(len(is_date)):
    if i == 0:
        if abs(boy[i] - girl[i]) <= 0.25:
            is_date[i] = 1
    else:
        is_date[i] = is_date[i - 1] + int((abs(boy[i] - girl[i]) <= 0.25))
```

```python
frequency = is_date / N
```

```python
plt.figure(figsize=(12, 4))

v = np.array([0.51] * len(N))
E = 0.05
plt.plot(N, frequency, color='red', marker='o')
plt.plot(N, v, color='orange')
plt.plot(N, v - E, color='blue')
plt.plot(N, v + E, color='blue')

plt.xlabel('N')
plt.ylabel('V')
plt.show()
```
