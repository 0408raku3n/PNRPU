```python
import matplotlib.pyplot as plt
import random as rd
import numpy as np
```

```python
h_all = [1, 1, 2, 3 , 4, 4, 2, 1, 1]
x = [13, 15, 17, 19, 21, 23, 25, 27, 29]

plt.figure(figsize=(10, 5))
plt.fill_between(x, h_all, color='skyblue')
plt.show()
```

```python
H = sum(h_all)
print(H)
```

```python
p_all = []
for h in h_all:
    p_all.append(h / H)
print(p_all, sum(p_all))
```

```python
interval = []
p_prev = 0
for p in p_all:
    interval.append(p_prev + p)
    p_prev = p_prev + p
print(interval)

plt.figure(figsize=(8, 3))

plt.plot(interval, len(interval) * [0], marker='x')

plt.show()
```

```python
interval_count = dict()
for inter in interval:
    interval_count[inter] = 0
for _ in range(900):
    number = rd.random()
    for key in interval_count.keys():
        if number <= key:
            interval_count[key] += 1
            break
print(interval_count)
```

```python
new_h = []
for key, item in interval_count.items():
    new_h.append(item)
print(new_h)
```

```python
plt.clf()

plt.fill_between(interval, new_h, color='skyblue')
plt.show()
```
