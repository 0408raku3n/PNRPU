### Начальные условия:
- A(0) = 0
- B(0) = 0
- C(0) = 200
- D(0) = 100
- E(0) = 0
- F(0) = 0
- G(0) = 0
- L(0) = 0

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
entities = ['A', 'B', 'C', 'D', 'E', 'U', 'F', 'G', 'L']
count_steps = 20
step = 0.5
A_values = np.array([0] * count_steps, dtype=float)
B_values = np.array([0] * count_steps, dtype=float)
C_values = np.array([0] * count_steps, dtype=float)
D_values = np.array([0] * count_steps, dtype=float)
E_values = np.array([0] * count_steps, dtype=float)
U_values = np.array([0] * count_steps, dtype=float)
F_values = np.array([0] * count_steps, dtype=float)
G_values = np.array([0] * count_steps, dtype=float)
L_values = np.array([0] * count_steps, dtype=float)
```

```python
A_values[0] = 10
B_values[0] = 10
C_values[0] = 30
D_values[0] = 15
E_values[0] = 5
F_values[0] = 10
G_values[0] = 10
L_values[0] = 0
U_values[0] = 50


k1 = 0.05
k2 = 0.2
k3 = 0.5
k4 = 0.3
k5 = 0.6
k6 = 0.1
k7 = 0.8
k8 = 1.2
k9 = 0.4
k10 = 0.5
k11 = 0.2
kL = 1.2
kU = 0.02
R = 10
```

```python
for i in range(1, count_steps):
    A = A_values[i - 1]
    B = B_values[i - 1]
    C = C_values[i - 1]
    D = D_values[i - 1]
    E = E_values[i - 1]
    U = U_values[i - 1]
    F = F_values[i - 1]
    G = G_values[i - 1]
    L = L_values[i - 1]

    dA = k1 * (B - A) + k2 * (C - A) + k3 * (D - A) + k4 * (E - A)
    dB = k1 * (A - B) + k5 * (C - B) + k6 * (G - B) + k7 * L
    dC = 0
    dD = 0
    dE = k4 * (A - E) + kU * U 
    dU = kU
    dF = k8 * (G - F) + k9 * (C - F) + k10 * (D - F)
    dG = k8 * (F - G) + k6 * (B - G) + k11 * (C - G)
    dL = kL

    A_values[i] = A + dA * step
    B_values[i] = B + dB * step
    C_values[i] = C + dC * step
    D_values[i] = D + dD * step
    E_values[i] = E + dE * step
    U_values[i] = U + dU * step
    F_values[i] = F + dF * step
    G_values[i] = G + dG * step
    L_values[i] = L + dL * step
```

```python
plt.figure(figsize=(8, 6))

x = np.arange(0, count_steps // 2, step)

plt.plot(x, A_values, label='A', color='green')
plt.plot(x, B_values, label='B', color='red')
plt.plot(x, C_values, label='C', color='skyblue')
plt.plot(x, D_values, label='D', color='green')
plt.plot(x, E_values, label='E', color='purple')
plt.plot(x, U_values, label='U', color='black')
plt.plot(x, F_values, label='F', color='blue')
plt.plot(x, G_values, label='G', color='grey')
plt.plot(x, L_values, label='L', color='cyan')


plt.legend()
plt.xlabel('t')
plt.ylabel('T')
plt.show()
```
